```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import bitstring
import matplotlib as mpl
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})
```

```python
plot_x, plot_y, plot_y_ps = np.loadtxt("text/iQPE_1st_CSS.txt", delimiter=",")

_, _ , plot_y_ps_X = np.loadtxt("text/iQPE_1st_CSS_X.txt",delimiter=",")

x,y = np.loadtxt("text/iQPE_1st.txt", delimiter=",")

x_1,y_1 = np.loadtxt("text/iQPE_1st_cz.txt", delimiter=",")

shots = 10000

err = [np.sqrt((i-i**2)/shots) for i in plot_x]

fig, ax = plt.subplots(figsize=(10,5))
ax.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 0))

ax.plot(plot_x,plot_y, label="Raw", marker = ".",markersize = 15, alpha = 0.5)
ax.plot(plot_x,plot_y_ps, label="Post selection", marker = ".",markersize = 15, alpha = 0.5)
ax.plot(plot_x,plot_y_ps_X, label="Post selection (with X)", marker = ".",markersize = 15, alpha = 0.5)
ax.plot(x,y, label="No encoding", marker = ".",markersize = 15, alpha = 0.5)
ax.plot(x_1,y_1, label="No encoding (CZ)", marker = ".",markersize = 15, alpha = 0.5)
ax = plt.errorbar(plot_x, plot_y, yerr=err, ecolor="red", linestyle="")
ax = plt.errorbar(plot_x, plot_y_ps, yerr=err, ecolor="red", linestyle="")
ax = plt.errorbar(x, y, yerr=err, ecolor="red", linestyle="")
ax = plt.errorbar(x_1, y_1, yerr=err, ecolor="red", linestyle="")
plt.title("1st iteration of iQPE", fontsize=25)
plt.xlabel("Pauli error probability",fontsize=25)
plt.ylabel("Successrate",fontsize=25)
plt.xticks(fontsize=25) 
plt.yticks(fontsize=25)
plt.grid()
plt.legend(fontsize=15)
plt.tight_layout()
plt.savefig("1.png", dpi = 200)
plt.show()
```

```python
data = np.loadtxt("text/RAW_Carbon_1st_Pauli_10k.txt", delimiter = ",")

data = np.delete(data, (0), axis = 0)

pre, post, nn, ne, en, ee, pre_QEC, post_QEC, nn_QEC, ne_QEC, en_QEC, ee_QEC = data[:,0], data[:,1], data[:,2], data[:,3], data[:,4], data[:,5], data[:,6], data[:,7], data[:,8], data[:,9], data[:,10], data[:,11]

fig, (ax1,ax2,ax3) = plt.subplots(figsize=(15,5),nrows=1,ncols=3)
# ax1.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 0))
# ax2.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 0))
# ax3.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals = 0))

x = np.linspace(0,0.1,20)

y = []
y_no_QEC = []
for i in range(len(x)):
    y.append(en_QEC[i]+ee_QEC[i])
    y_no_QEC.append(en[i]+ee[i])

y_PS = []
y_no_QEC_PS = []
for i in range(len(x)):
    if en_QEC[i]+ee_QEC[i]+ne_QEC[i]+nn_QEC[i] != 0:
        y_PS.append((en_QEC[i]+ee_QEC[i])/(en_QEC[i]+ee_QEC[i]+ne_QEC[i]+nn_QEC[i]))
    else:
        y_PS.append(0.5)
    if en[i]+ee[i]+nn[i]+ne[i] != 0:
        y_no_QEC_PS.append((en[i]+ee[i])/(en[i]+ee[i]+nn[i]+ne[i]))
    else:
        y_no_QEC_PS.append(0.5)


ax1.plot(x,y, label="With QEC (No PS)", marker = ".",markersize = 15, alpha = 0.5)
ax1.plot(x,y_no_QEC, label="No QEC (No PS)", marker = ".",markersize = 15, alpha = 0.5)

ax2.plot(x,y_PS, label="With QEC", marker = ".",markersize = 15, alpha = 0.5)
ax2.plot(x,y_no_QEC_PS, label="No QEC", marker = ".",markersize = 15, alpha = 0.5)

ax3.plot(x, pre_QEC, label="Preselection (QEC)", marker = ".",markersize = 15, alpha = 0.5)
ax3.plot(x, pre, label="Preselection (No QEC)", marker = ".",markersize = 15, alpha = 0.5)

ax3.plot(x, post_QEC, label="Postselection (QEC)", marker = ".",markersize = 15, alpha = 0.5)
ax3.plot(x, post, label="Postselection (No QEC)", marker = ".",markersize = 15, alpha = 0.5)

#ax1.set_title("1st iteration of iQPE on the [12, 2, 4] Carbon Code", fontsize=20)
ax1.set_title("First iteration", fontsize=20)
ax2.set_title("With Pre-/Postselection", fontsize=20)
ax3.set_title("Pre-/Postselectionrates", fontsize=20)
ax1.set_xlabel("Pauli Error probability", fontsize=20)
ax2.set_xlabel("Pauli Error probability", fontsize=20)
ax3.set_xlabel("Pauli Error probability", fontsize=20)
ax1.set_ylabel("Successrate", fontsize=20)
ax2.set_ylabel("Successrate", fontsize=20)
ax3.set_ylabel("Discardrate", fontsize=20)
ax1.tick_params(axis='both', which='major', labelsize=20)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax3.tick_params(axis='both', which='major', labelsize=20)
ax1.grid(), ax2.grid(), ax3.grid()
ax1.legend(fontsize=15)
ax2.legend(fontsize=15)
ax3.legend(fontsize=13)
plt.tight_layout()
plt.savefig("9.png", dpi = 200)
plt.show()
```
